```python
#drop rows with empty rain accumulation and rain duration
null_rain_duration = data[ data['rain_duration']== 0 ]
data = data.drop(null_rain_duration.index, axis=0)

null_rain_duration = data[ data['rain_duration']== 0 ]
data = data.drop(null_rain_duration.index, axis=0)

data
```

```python
#data preparation
#here we will use only those meaningful columns
new_col=['air_pressure','air_temp','avg_wind_direction','avg_wind_speed','max_wind_direction','max_wind_speed','relative_humidity']
new_data=data.loc[:,new_col]
new_data.head()
```

```python
#data preprocessing 

#this method to drop all nan values before preprocessing, else preporcessing method throws an error
for i in new_col:
    new_data = new_data[np.isfinite(new_data[i])]
    

from sklearn.preprocessing import StandardScaler
StandardScaler(copy=True, with_mean=True, with_std=True)
scaler = StandardScaler()
print(scaler.fit(new_data))
print(scaler.mean_)
new=scaler.fit_transform(new_data)
#Standardize features by removing the mean and scaling to unit variance
#Centering and scaling happen independently on each feature by computing the relevant statistics on the samples
#in the training set. 
#Mean and standard deviation are then stored to be used on later data using the transform method.
```

```python
#convert array to a dataframe 
scaled_data=pd.DataFrame(new,index=new_data.index, columns=new_data.columns)
```

```python
#Modeling

#In our example (if we didn't have access to the k-means algorithm),
#it would be the same as guessing that each weather group would have certain rain accumulation, relative humidity,air pressure,etc,
#with multiple tests and experiments. However, using the K-means clustering we can do all this process much easier.
kmeans = KMeans(n_clusters=12)  
kmeans.fit(scaled_data)
```

```python
#print clusters centroids
print(kmeans.cluster_centers_)
```

```python
#print labels of each centroid
labels=kmeans.labels_
print(labels)
```

```python
#display cluster label of each record
scaled_data['cluster']=labels
scaled_data.head()
```

```python
#We can easily check the centroid values by averaging the features in each cluster.
scaled_data.groupby('cluster').mean()
```
